```python
from __future__ import print_function
import antlr4
import copy
import json
import re
# compiler
from pymola.generated.ModelicaLexer import ModelicaLexer
from pymola.generated.ModelicaParser import ModelicaParser
from pymola.generated.ModelicaListener import ModelicaListener
```

```python
%%file RigidBody.mo
model Accelerometer
    input Real a_x;
    input Real a_y;
    input Real a_z;
    output Real m_a_x;
    output Real m_a_y;
    output Real m_a_z;
    parameter Real b_x = 1e-5 "bias x";
    parameter Real b_y = 1e-5 "bias y";
    parameter Real b_z = 1e-5 "bias z";
equation
    m_a_x = a_x + bias_x;
    m_a_y = a_y + bias_y;
    m_a_z = a_z + bias_z;
end Accel;

model RigidBody "body"
    input Real F_x "force x";
    input Real F_y "force y";
    input Real F_z "force z";
    input Real M_x "moment x";
    input Real M_y "moment y";
    input Real M_z "moment z";
    parameter Real m = 1 "mass";
    parameter Real Jx = 1 "Jx";
    parameter Real Jy = 1 "Jy";
    parameter Real Jz = 1 "Jz";
    output Real x, y, z;
    output Real v_x, v_y, v_z;
    output Real a_x, a_y, a_z;
    output Real q0, q1, q2, q3;
    output Real w_x, w_y, w_z;
    output Real alpha_x, alpha_y, alpha_z;
equation
    // translation
    der(x) = v_x;
    der(y) = v_y;
    der(z) = v_z;
    der(v_x) = a_x;
    der(v_y) = a_y;
    der(v_z) = a_z;
    m*a_x = F_x;
    m*a_y = F_y;
    m*a_z = F_z;
    // rotation
    der(q0) = w_x;
    der(q1) = w_y;
    der(q2) = w_z;
    der(q3) = w_z;
    der(w_x) = alpha_x;
    der(w_y) = alpha_y;
    der(w_z) = alpha_z;
    J_x*alpha_x = M_x;
    J_y*alpha_y = M_y;
    J_z*alpha_z = M_z;
end RigidBody;

model Aircraft
    RigidBody body;
    Accelerometer accel;
equation
    connect(body.a_x, accel.a_x);
    connect(body.a_y, accel.a_y);
    connect(body.a_z, accel.a_z);
end Aircraft;
```

```python
modelica_model = open('RigidBody.mo', 'r').read()
input_stream = antlr4.InputStream(modelica_model)
lexer = ModelicaLexer(input_stream)
stream = antlr4.CommonTokenStream(lexer)
parser = ModelicaParser(stream)
tree = parser.stored_definition()
```

```python
class DaeCreator(ModelicaListener):

    def __init__(self, parser, trace):
        """
        Constructor
        """
        self._parser = parser
        self._trace = trace
        # variable to hold current model
        # final output model
        self.c_file = None
        self.c_class = None
        self.c_eq_list = None
        self.c_comp_list = None
        self.result = None
        self._val_dict = {}
        self._parser = parser
        self._trace = trace

    @staticmethod
    def print_indented(ldr, s):
        if s is not None:
            for line in str(s).split('\n'):
                print(ldr, line)

    def setValue(self, ctx, val):
        """
        Sets tree values.
        """
        if ctx.depth() == 1:
            self.result = val
        else:
            self._val_dict[ctx] = val

    def getValue(self, ctx):
        """
        Gets tree values.
        """
        assert ctx is not None
        if ctx.depth() == 1:
            return self.result
        else:
            try:
                return self._val_dict[ctx]
            except KeyError:
                return None

    def enterEveryRule(self, ctx):
        if self._trace:
            ldr = " "*(ctx.depth()-1)
            rule = self._parser.ruleNames[ctx.getRuleIndex()]

            print(ldr, rule + "{")
            in_str = ctx.getText()
            if in_str > 0:
                print(ldr, "==============================")
                print(ldr, "INPUT\n")
                self.print_indented(ldr, ctx.getText())
                print(ldr, "==============================\n")

    def visitTerminal(self, node):
        pass

    def visitErrorNode(self, node):
        pass

    def exitEveryRule(self, ctx):
        rule = self._parser.ruleNames[ctx.getRuleIndex()]
        if self._trace:
            ldr = " "*ctx.depth()
            lines = None
            try:
                lines = self.getValue(ctx)
            except KeyError as e:
                pass

            if lines is not None:
                print(ldr, "==============================")
                print(ldr, "OUTPUT\n")
                self.print_indented(ldr, lines)
                print(ldr, "==============================\n")

            print(ldr + '} //' + rule + '\n')
        #if self.getValue(ctx) is None:
        #    raise RuntimeError("no value set for {:s}:\ninput:\n{:s}".format(rule, ctx.getText()))

        
    # stored definition
    def enterStored_definition(self, ctx):
        self.c_file = {
            'classes': {}
        }

    def exitStored_definition(self, ctx):
        self.result = self.c_file
        self.c_file = None

    # class
    def enterStored_definition_class(self, ctx):
        self.c_class =  {
            'comment': [],
            'diff_eqs': [],
            'init_eqs': [],
            'alg_eqs': [],
            'states': [],
            'inputs': [],
            'outputs': [],
            'symbols': [],
            'parameters': {},
        }
        
    def exitStored_definition_class(self, ctx):
        cls = self.c_class
        spec = ctx.class_definition().class_specifier()
        name = spec.IDENT()[0].getText()
        cls['comment'] = spec.string_comment().getText()[1:-1]
        self.c_file['classes'][name] = cls
        self.c_class = None
    
    # components
    def enterComponent_clause(self, ctx):
        self.c_comp_list = []
        
    def exitComponent_clause(self, ctx):
        type_prefix = ctx.type_prefix().getText() 
        for comp in self.c_comp_list:
            decl = comp.declaration()
            name = decl.IDENT().getText()
            comment = comp.comment().getText()[1:-1]
            try:
                val = float(decl.modification().expression().getText())
            except:
                val = 0
            if type_prefix == 'parameter' and name not in self.c_class['parameters'].keys():
                self.c_class['parameters'][name] = {
                    'value': val,
                    'comment': comment
                }
            elif type_prefix == 'input' and name not in self.c_class['inputs']:
                self.c_class['inputs'] += [name]
            elif type_prefix == 'output' and name not in self.c_class['outputs']:
                self.c_class['outputs'] += [name]
            elif name not in self.c_class['symbols']:
                self.c_class['symbols'] += [name]
        self.c_comp_list = None

    def exitComponent_declaration(self, ctx):
        self.c_comp_list += [ctx]

    # equation
    def enterEquation_section(self, ctx):
        self.c_eq_list = []
        
    def exitEquation_section(self, ctx):
        for i in range(len(self.c_eq_list)):
            txt = self.c_eq_list[i]
            # make states functions of time
            for state in set(self.c_class['states'] + self.c_class['inputs'] + self.c_class['outputs']):
                txt = re.sub(
                    r'(^|[^\w])({:s})'.format(state),
                    r'\g<1>\g<2>(t)', txt)
            # convert derivative
            txt = re.sub(r'der\(([^\)]+)\)', r'(\g<1>).diff(t)', txt)
            self.c_eq_list[i] = txt
            
        if ctx.INITIAL() != None:
            self.c_class['initial_eqs'] += self.c_eq_list
        else:
            self.c_class['diff_eqs'] += self.c_eq_list
        self.c_eq_list = None
        
    def exitEquation_simple(self, ctx):
        self.c_eq_list += ['{:s} - ({:s})'.format(
            ctx.simple_expression().getText(),
            ctx.expression().getText())]
    
    # derivative
    def exitPrimary_derivative(self, ctx):
        args = ctx.function_call_args().function_arguments()
        state = args.getText()
        self.c_class['states'] += [state]
        if state in self.c_class['symbols']:
            self.c_class['symbols'].remove(state)
   
    def sympy_eq_export(self, class_name):
        import sympy
        t = sympy.Symbol('t')
        cls = self.result['classes'][class_name]
        eqs = sympy.Matrix(sympy.sympify(cls['diff_eqs']))
        x = sympy.Matrix(sympy.sympify(['{:s}(t)'.format(e) for e in cls['states']]))
        u = sympy.Matrix(sympy.sympify(['{:s}(t)'.format(e) for e in cls['inputs']]))
        y  = sympy.Matrix(sympy.sympify(['{:s}(t)'.format(e) for e in cls['outputs']]))
        p  = sympy.Matrix(sympy.sympify(['{:s}'.format(e) for e in cls['parameters']]))
        fg_sol = sympy.solve(eqs, list(x.diff(t)) + list(y))
        f = x.diff(t).subs(fg_sol)
        g = y.subs(fg_sol)
        return {'t': t, 'x': x, 'u': u, 'y': y, 'p': p, 'f': f, 'g': g}
```

```python
ast = DaeCreator(parser, trace=False)
walker = antlr4.ParseTreeWalker()
walker.walk(ast, tree)
```

```python
print(json.dumps(ast.result, indent=2))
```
